```python
#...............REPLACE NAN VALUES..................................
columns_to_fill = ['people_fully_vaccinated', 'daily_vaccinations_raw','daily_vaccinations','total_vaccinations_per_hundred','people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred','daily_vaccinations_per_million','total_vaccinations','people_vaccinated']
data[columns_to_fill] = data[columns_to_fill].apply(lambda col: col.fillna(col.mean()))


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Step 1: Split the 'vaccines' column and create a frequency table
# Split the string in 'vaccines' and flatten the list
vaccines_split = data['vaccines'].str.split(', ')
# Unlist and create a frequency table
data_used = vaccines_split.explode().value_counts().reset_index()

# Step 2: Rename the columns
data_used.columns = ['data', 'Freq']

# Display the frequency table
print(data_used)

# Step 3: Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Freq', y='data', data=data_used, palette='viridis')

# Customize the plot
plt.title('Type of COVID-19 Vaccines Used in the World')
plt.xlabel('Number of Countries')
plt.ylabel('Vaccines Used')

# Show the plot
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_country = (
    data.groupby('country')
    .agg(percentage_of_people_fully_vaccinated=('people_fully_vaccinated_per_hundred', 'max'))
    .reset_index()
    .sort_values(by='percentage_of_people_fully_vaccinated', ascending=False)
)


top_10_countries = data_country.head(20)


plt.figure(figsize=(10, 6))
sns.barplot(x='percentage_of_people_fully_vaccinated',
            y='country',
            data=top_10_countries,
            palette='viridis')


plt.title('10 Countries with The Highest Percentage of People Fully Vaccinated')
plt.xlabel('Percentage of People Fully Vaccinated (%)')
plt.ylabel('Country')


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Define a function to remove outliers based on IQR
def remove_outliers_iqr(df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1
    # Filter the dataframe to remove outliers
    df_cleaned = df[~((df[columns] < (Q1 - 1.5 * IQR)) | (df[columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df_cleaned

# Select the relevant columns for outlier removal
columns = ['total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'daily_vaccinations']

# Remove outliers
data = remove_outliers_iqr(data, columns)

# Check shape of the original vs cleaned data
print(f"Original dataset shape: {data.shape}")
print(f"Dataset shape after removing outliers: {data.shape}")

# Visualize cleaned data
plt.figure(figsize=(10, 6))
sns.boxplot(data[columns])  # Use cleaned_data for visualization
plt.title("Boxplot of Vaccination Data (Outliers Removed)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


features = data[['people_vaccinated', 'people_fully_vaccinated', 'daily_vaccinations']]  # Add more features as needed
target = data['total_vaccinations']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt

# Plotting actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # Line for perfect prediction
plt.title("Actual vs Predicted Values")
plt.xlabel("Actual Total Vaccinations")
plt.ylabel("Predicted Total Vaccinations")
plt.grid()
plt.show()
```
